## 判断文件类型

```python
file_name = input('please input a file name')
file_type = file_name.split('.')[1]
if file_type == 'mp4':
    print('.mp4 video file')
elif file_type == 'mp3':
    print('.mp3 audio file')
elif file_type  == 'txt':
    print('.txt text file')
elif file_type == 'doc':
    print('.doc Microsoft document file')
elif file_type == 'py':
    print('.py python code file')
else:
    print('not recognized')

if file_name.endswith('.mp4'):
    print('it\'s a .mp4 file')
else:
    print('not a .mp4 file')
```

## 解析目录并显示图片

```python
import os

root_dir = './data/my_data/'
id_file = 'my_id.txt'
img_path = os.path.join(root_dir, 'my_img')

img_files = []
with open(os.path.join(root_dir, id_file)) as f:
    img_files = [os.path.join(img_path,line[:-1]) for line in f.readlines()]
img_files
```

```python
import matplotlib.pyplot as plt
import random

random_index = random.randrange(0, len(img_files))
img_path = img_files[random_index].split()[0]

image = plt.imread(img_path)
plt.imshow(image)
plt.show()
```
