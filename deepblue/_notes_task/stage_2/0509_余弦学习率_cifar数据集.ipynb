# 余弦退火学习率+周期性重启

$\eta_t = \eta_min+\frac{1}{2}(\eta_{max}-\eta_{min})(1+cos(\frac{T_{cur}}{T_i}\pi)$

```python
import numpy as np
import matplotlib.pyplot as plt

def lr_schedule_cosine(lr_min, lr_max, per_epochs):
    def compute(epoch):
        return lr_min + 0.5 * (lr_max - lr_min) * (1 + np.cos(epoch / per_epochs * np.pi))
    return compute

x = np.arange(0, 300)
y = lr_schedule_cosine(1e-2, 1e-1, 300)(x)
plt.plot(x, y)

x2 = list(range(100))+list(range(100))+list(range(100))
y2 = list(map(lr_schedule_cosine(1e-2, 1e-1, 100), x2))
plt.plot(x, y2, 'r:')
```
